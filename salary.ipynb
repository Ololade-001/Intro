## I'm checking for the total number of job titles that contain captain

```python
captain_jobs = sal['JobTitle'].str.contains('captain', case=False, na=False)

# Count how many job titles contain "captain"
total_captain_jobs = captain_jobs.sum()

print(f"Total number of job titles containing 'captain': {total_captain_jobs}")
```

## Here I'm calculate for the minimum, maximum and average Basepay 

```python
# Calculate the minimum, maximum, and average BasePay
min_base_pay = sal['BasePay'].min()
max_base_pay = sal['BasePay'].max()
average_base_pay = sal['BasePay'].mean()

# Display the results
print(f"Minimum BasePay: {min_base_pay}")
print(f"Maximum BasePay: {max_base_pay}")
print(f"Average BasePay: {average_base_pay}")
```

## Here I'm replacing the not provided with NaN in the employeeName column

```python
# Replace 'Not provided' with NaN in the EmployeeName column
sal['EmployeeName'] = sal['EmployeeName'].replace('Not provided', np.nan)

# Optionally, display the updated DataFrame
print(sal)
```

## Drop rows with five missing value 

```python
# Drop rows with exactly 5 missing values
sal = sal[sal.isna().sum(axis=1) != 5]
sal
```

## I want to know the job title of Albert pardini 

```python
# Filter the rows where the Name is 'Albert Pardini'
job_title = sal[sal['EmployeeName'] == 'Albert Pardini']['JobTitle'].values

# If you want to print or access the job title
if len(job_title) > 0:
    print(f"The job title of Albert Pardini is: {job_title[0]}")
else:
    print("Albert Pardini not found in the dataset")
```

## I want to know how much Albert Pardini makes including benefits

```python
# Filter for Albert Pardini's row
albert_data = sal[sal['EmployeeName'] == 'Albert Pardini']

# Check if the record exists and calculate total earnings (Salary + Benefits)
if not albert_data.empty:
    total_earnings = albert_data['TotalPay'].values[0] + albert_data['TotalPayBenefits'].values[0]
    print(f"Albert Pardini makes ${total_earnings} including benefits.")
else:
    print("Albert Pardini not found in the dataset.")
```

## Here i want to know the person with the highest basepay

```python
# Find the index of the person with the highest base pay
highest_basepay_index = sal['BasePay'].idxmax()

# Retrieve the name of the person with the highest basepay
person_with_highest_basepay = sal.loc[highest_basepay_index, 'EmployeeName']

# Print the name
print(f"The person with the highest basepay is: {person_with_highest_basepay}")
```

## I want to know the average Basepay of all Employees per Year

```python
# Group by 'Year' and calculate the mean of 'Basepay'
average_basepay_per_year = sal.groupby('Year')['BasePay'].mean()
# Convert to DataFrame if needed
average_basepay_per_year = sal.groupby('Year', as_index=False)['BasePay'].mean()
print(average_basepay_per_year)
```

## I want to know the Average Basepay Of All Employee Per Job Title

```python
average_basepay_per_job_title = sal.groupby('JobTitle')['BasePay'].mean()
average_basepay_per_job_title = sal.groupby('JobTitle', as_index=False)['BasePay'].mean()
# Print the average base pay for each job title
print(average_basepay_per_job_title)
```

## I want to know the Average Basepay of All Employee Having job Title Accountant

```python
# Filter the dataset for rows where 'Job Title' is 'Accountant'
accountants = sal[sal['JobTitle'] == 'ACCOUNTANT']
# Calculate the average base pay for accountants
average_basepay_accountant = accountants['BasePay'].mean()
# Print the result
print(f"The average base pay of accountants is: ${average_basepay_accountant}")
```
